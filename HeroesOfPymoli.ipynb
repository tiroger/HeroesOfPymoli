### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* On average, female players spend as much as male players - $4.47 vs 4.07, respectively. But 80% of top spenders are males

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Load files
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Calculating total number of players based on the number unique screen name(SN)
player_num = len(purchase_data['SN'].unique())

# Creating a new table to contain the total number of players player
player_count = {"Player Count" : [player_num]}

# Storing data into a new dataframe
player_count_df = pd.DataFrame(player_count)
player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculating the number of unique items purchased based on the item name
unique_items = purchase_data['Item Name'].nunique()

# Calculating the average price by finding the mean of the purchase prices
avg_price = purchase_data['Price'].mean()

# Calculating the total number of purchases
purchase = purchase_data['Purchase ID'].count()

# Calculating the total revenue by summing all the prices
revenue = purchase_data['Price'].sum()

# Storing data into a new dataframe
purchase_summary = {'Number of Unique Items': [unique_items], 
                    'Average Price': [avg_price], 
                    'Number of Purchases': [purchase], 
                    'Total Revenue': [revenue]
                   }

purchase_summary_df = pd.DataFrame(purchase_summary)

# Formatting columns data into a new dataframe
purchase_summary_df['Average Price'] = purchase_summary_df['Average Price'].map('${:.2f}'.format)
purchase_summary_df['Total Revenue'] = purchase_summary_df['Total Revenue'].map('${:,.2f}'.format)

purchase_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# In order to count the genders we remove the duplicate screen names (SN)
unique_player_gender = purchase_data.drop_duplicates(subset='SN', keep="last")

# Grouping the unique players by gender
grouped_unique_player_gender = unique_player_gender.groupby('Gender')

# Calculating the number of males, females, and others in the grouped data by counting screen names
gen_dem = grouped_unique_player_gender['SN'].count()

# Calculating the percentage of genders over the total number of players
gen_perc = gen_dem / player_num * 100

# Creating new dataframe with values above
dem_sum = {'Total Count' : gen_dem, 
           'Percentage of Player': gen_perc
          }

dem_sum_df = pd.DataFrame(dem_sum)

# Formatting dataframe
dem_sum_df['Percentage of Player'] = dem_sum_df['Percentage of Player'].map('{:,.1f}'.format)
dem_sum_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Finding max age
max_age = purchase_data['Age'].max()

# Bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
group_names = ["< 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# Binning the values based off age groups
purchase_data['Age Groups'] = pd.cut(purchase_data['Age'], bins, labels = group_names)

# Grouping the data based on created bins (age groups)
binned_purchase_data = purchase_data.groupby('Age Groups')

# Calculating the number of players in each age group
age_total = binned_purchase_data['SN'].nunique()

# Calculating the percentage of each group relative to the total number of players
age_perc = age_total / player_num * 100

# Creating a new dataframe to hold the values above
age_dem = {'Total Count': age_total, "Percentage of Players": age_perc}
age_dem_df = pd.DataFrame(age_dem)

# Formating numbers
pd.options.display.float_format = '{:.1f}'.format

age_dem_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Grouping data by screen name
grouped_player = purchase_data.groupby('SN')

# Calculating the number of purchases per player
player_purch = grouped_player['Purchase ID'].count()

# Finding the player gender
player_gen = grouped_player['Gender'].unique()

# Calculating the average purchase per player
player_avg = grouped_player['Price'].mean()

# Calculating the total purchase value per player
player_total = grouped_player['Price'].sum()

# Creating a new dataframe to hold the above values
player_summary = {'Gender': player_gen, 
                  'Purchase Count': player_purch, 
                  'Average Purchase Price': player_avg, 
                  'Total Purchase Value': player_total
                 }

player_summary_df = pd.DataFrame(player_summary)

# Sorting in descending order to find the top spenders
top_spenders = player_summary_df.sort_values('Total Purchase Value', ascending=False)

# Removing brackets from Gender
top_spenders['Gender'] = top_spenders['Gender'].str[0]
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Grouping data by item number and item name
grouped_item = purchase_data.groupby(['Item ID', 'Item Name'])

# Calculating the number of times each item was purchased
item_counts = grouped_item['Item ID'].count()

# Finding the price of the item
item_price = grouped_item['Price'].unique()

# Finding the total amout for each item
item_total = grouped_item['Price'].sum()

item_summary = {'Purchase Count': item_counts, 
                'Item Price': item_price, 
                'Total Purchase Value': item_total
               }
# Creating dataframe to store above values
item_summary_df = pd.DataFrame(item_summary)

# Formatting columns
item_summary_df['Item Price'] = item_summary_df['Item Price'].str[0]

item_summary_df = item_summary_df.sort_values('Purchase Count', ascending=False)
```
